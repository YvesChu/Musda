```python
def mid2np(path):
  muspy_Music=muspy.read_midi(path)       #to muspy.Music
  numpy=muspy.to_note_representation(muspy_Music)#to numpy
  return numpy



def np2mid(path_gendir,numpy,sidx):
  muspy_Music=muspy.from_note_representation(numpy,resolution=275)#to muspy.Music
  # path_gendir_add = os.path.join(path_gendir,str(times))
  # os.mkdir(path_gendir_add)
  # path_outfiles = os.path.join(path_gendir_add,'get_{}.mid'.format(str(sidx)))
  path_outfiles = os.path.join(path_gendir,'get_{}.mid'.format(str(sidx)))
  muspy.write_midi(path_outfiles,muspy_Music)        #to mid
```

```python
def rand_del(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range(0,line_num):
    prob = random.random()
    if prob < 0.1:
      numpy=np.delete(numpy,obj,axis=0)
  return numpy

def rand_add_3rd(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.02:
      arr=numpy[obj]
      if arr[1]+4>108:
        continue
      numpy = np.insert(numpy,obj,[arr[0],arr[1]+4,arr[2],arr[3]],axis=0)
  return numpy

def rand_add_5th(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.02:
      arr=numpy[obj]
      if arr[1]+7>108:
        continue
      numpy = np.insert(numpy,obj,[arr[0],arr[1]+7,arr[2],arr[3]],axis=0)
  return numpy

def rand_add_Oct(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.05:
      arr=numpy[obj]
      if arr[1]+12>108:
        continue
      numpy = np.insert(numpy,obj,[arr[0],arr[1]+12,arr[2],arr[3]],axis=0)
  return numpy

def pitch_shift_3rd(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.02:
      arr=numpy[obj]
      arr[1]+=4
      if arr[1]>108:
        arr[1]-=4
  return numpy

def pitch_shift_5th(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.02:
      arr=numpy[obj]
      arr[1]+=7
      if arr[1]>108:
        arr[1]-=7
  return numpy

def pitch_shift_Oct(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range (0,line_num):
    prob=random.random()
    if prob < 0.05:
      arr=numpy[obj]
      arr[1]+=12
      if arr[1]>108:
        arr[1]-=12
  return numpy
```

```python
def velocity_float(numpy):
  line_num = np.shape(numpy)[0]
  for obj in range(0,line_num):
    arr=numpy[obj]
    float_parameter=random.randint(-0.5,0.5)
    arr[3]+=float_parameter
  return numpy

def duration_float(numpy):
  line_num=np.shape(numpy)[0]
  for obj in range(0,line_num):
    arr=numpy[obj]
    float_parameter=random.randint(-0.5,0.5)
    arr[2]+=float_parameter
  return numpy 
```

```python
def key_shift(numpy):
  line_num = np.shape(numpy)[0]
  key_shift=random.randint(-12,12)
  key_max=np.argmin(numpy[:,1])
  for obj in range(0,line_num):
    arr=numpy[obj]
    arr[1]+=key_shift
    if arr[1]>108:
      arr[1]=108
  return numpy
```
